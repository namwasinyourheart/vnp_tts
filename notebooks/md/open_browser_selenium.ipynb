```python
def _get_chrome_options(headless: bool = False, profile: str = None, guess: bool = False) -> uc.ChromeOptions:
    """Creates a fresh ChromeOptions object."""
    options = uc.ChromeOptions()
    if headless:
        options.add_argument("--headless=new")
    options.add_argument("--disable-gpu")
    options.add_argument("--no-sandbox")
    
    # Profile handling
    if not guess and profile and profile != "default":
        profile_path = f"/home/nampv1/selenium-chrome-profiles/{profile}"
        print(f"Using Chrome profile from: {profile_path}")
        options.add_argument(f"--user-data-dir={profile_path}")
        options.add_argument(f"--profile-directory={profile}")
    else:
        print("Running in guess mode or with default profile.")
    
    return options

def _create_driver(headless: bool = False, profile: str = None, guess: bool = False):
    """Helper function to create a new WebDriver instance with retry logic."""
    import time
    max_retries = 3
    for attempt in range(max_retries):
        try:
            options = _get_chrome_options(headless, profile, guess)
            driver = uc.Chrome(options=options, version_main=142)
            return driver
        except Exception as e:
            logger.error(f"WebDriver creation failed on attempt {attempt + 1}: {e}")
            if attempt < max_retries - 1:
                time.sleep(5)  # Wait before retrying
            else:
                raise  # Re-raise the last exception if all retries fail
```

```python
driver = _create_driver()
```
